```python

def cluster_program(program):
    if 'FSD MERN' in program:
        return 'Full Stack Development'
    elif 'Android' in program or 'Flutter' in program:
        return 'Mobile Development'
    elif 'Cyber Security' in program:
        return 'Cyber Security'
    elif 'AWS' in program:
        return 'Cloud Computing'
    elif 'Python' in program:
        return 'Programming'
    elif 'ML/AI' in program:
        return 'Machine Learning/AI'
    else:
        return program
```

```python
ict1['Program Cluster'] = ict1['Select Your Internship Program'].apply(cluster_program)
```

```python
ict1
```

```python
def clusterrating(rating):
    if rating == 5:
        return 'Excellent'
    elif rating == 4:
        return 'Good'
    elif rating == 3:
        return 'Average'
    elif rating == 2:
        return 'Below Average'
    elif rating == 1:
        return 'Poor'
    else:
        return 'Unknown'
```

```python
ict1.columns
```

```python
rating = ict1[["Today's grooming session was relevant and helpful for me to relate. ","Today's delivery of the content was clear and understandable",
       "How would you rate today's instructor?",
       "How would you rate today's coordinator support?",]]
```

```python
for i in rating:
  ict1[i]=ict1[i].apply(clusterrating)
```
