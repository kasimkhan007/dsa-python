```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
store=pd.read_excel("/content/Sample - Superstore.xls")
```

```python
ss=store
ss
```

```python
ss[['Sales', 'Profit']].hist()
plt.show()
```

```python
plt.scatter(ss['Sales'], ss['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs Profit')
plt.show()
```

```python
ss['Sales'].nlargest(10).plot.bar()
plt.xlabel('City')
plt.ylabel('Sales')
plt.title('Top 10 Cities by Sales')
plt.show()
```

```python
sns.boxplot(x='Category', y='Sales', data=ss)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales Distribution Across Categories')
plt.show()
```

```python
sns.pairplot(ss[['Sales', 'Profit', 'Quantity', 'Discount']])
plt.show()
```
