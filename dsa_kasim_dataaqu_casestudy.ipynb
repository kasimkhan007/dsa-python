Q6: Identify any potential outliers in a numerical column of your choice.


```python
q1=hd['price'].quantile(0.25)
q3=hd['price'].quantile(0.75)
```

```python
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3-1.5*iqr
outliner=hd[(hd['price']<lower_bound)] | hd[(hd['price']>upper_bound)]
```

```python
outliner
```

```python
outliner.sum()
```
